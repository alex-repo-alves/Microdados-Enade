```python
import plotly.graph_objects as go

fig = go.Figure()
fig.add_trace(go.Box(
    y=engenharias_ufpa_csv['NOTA PROVA'],
    name='Nota da Prova',
    marker_color= 'darkblue',
    boxmean='sd' # represent mean and standard deviation
))

fig.add_trace(go.Box(
    y=engenharias_ufpa_csv['NOTA FORMACAO GERAL'],
    name='Nota Formação Geral',
    marker_color= 'royalblue',
    boxmean='sd' # represent mean and standard deviation
))

fig.add_trace(go.Box(
    y=engenharias_ufpa_csv['NOTA PROVA OBJETIVA FORMACAO GERAL'],
    name='Nota Prova Objetiva Formação Geral',
    marker_color= 'royalblue',
    boxmean='sd' # represent mean and standard deviation
))

fig.add_trace(go.Box(
    y=engenharias_ufpa_csv['NOTA PROVA DISCURSIVA FORMACAO GERAL'],
    name='Nota Prova Discursiva Formação Geral',
    marker_color= 'royalblue',
    boxmean='sd' # represent mean and standard deviation
))

fig.add_trace(go.Box(
    y=engenharias_ufpa_csv['NOTA COMPONENTE ESPECIFICO'],
    name='Nota Componente Específico',
    marker_color= 'indianred',
    boxmean='sd' # represent mean and standard deviation
))

fig.add_trace(go.Box(
    y=engenharias_ufpa_csv['NOTA PROVA OBJETIVA COMPONENTE ESPECIFICO'],
    name='Nota Prova Objetiva Componente Específico',
    marker_color= 'indianred',
    boxmean='sd' # represent mean and standard deviation
))

fig.add_trace(go.Box(
    y=engenharias_ufpa_csv['NOTA PROVA DISCURSIVA COMPONENTE ESPECIFICO'],
    name='Nota Prova Discursiva Componente Específico',
    marker_color= 'indianred',
    boxmean='sd' # represent mean and standard deviation
))

fig.show()
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objs as go

df = engenharias_ufpa_csv
if isinstance(df, (pd.DatetimeIndex, pd.MultiIndex)):
	df = df.to_frame(index=False)

# remove any pre-existing indices for ease of use in the D-Tale code, but this is not required
df = df.reset_index().drop('index', axis=1, errors='ignore')
df.columns = [str(c) for c in df.columns]  # update columns to strings in case they are numbers

s = df[~pd.isnull(df['1 - Qual o grau de dificuldade desta prova na parte de Formação Geral?'])]['1 - Qual o grau de dificuldade desta prova na parte de Formação Geral?']
chart = pd.value_counts(s).to_frame(name='data')
ordinal_data = df.groupby('1 - Qual o grau de dificuldade desta prova na parte de Formação Geral?')[['NOTA FORMACAO GERAL']].mean()
chart['ordinal'] = ordinal_data
chart.index.name = 'labels'
chart = chart.reset_index().sort_values('ordinal')
chart = chart[:6]
charts = [go.Bar(x=chart['labels'].values, y=chart['data'].values, name='Frequency')]
charts.append(go.Scatter(
	x=chart['labels'].values, y=chart['ordinal'].values, yaxis='y2',
	name='NOTA FORMACAO GERAL (mean)', line={'shape': 'spline', 'smoothing': 0.3}, mode='lines'
))
figure = go.Figure(data=charts, layout=go.Layout({
    'barmode': 'group',
    'legend': {'orientation': 'h'},
    'title': {'text': '1 - Qual o grau de dificuldade desta prova na parte de Formação Geral? Value Counts'},
    'xaxis': {'title': {'text': '1 - Qual o grau de dificuldade desta prova na parte de Formação Geral?'}},
    'yaxis': {'title': {'text': 'Frequency'}},
    'yaxis2': {'overlaying': 'y', 'side': 'right', 'title': {'text': 'NOTA FORMACAO GERAL (mean)'}}
}))

figure.show()
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objs as go

df = engenharias_ufpa_csv
if isinstance(df, (pd.DatetimeIndex, pd.MultiIndex)):
	df = df.to_frame(index=False)

# remove any pre-existing indices for ease of use in the D-Tale code, but this is not required
df = df.reset_index().drop('index', axis=1, errors='ignore')
df.columns = [str(c) for c in df.columns]  # update columns to strings in case they are numbers

s = df[~pd.isnull(df['2 - Qual o grau de dificuldade desta prova na parte do Componente Específico?'])]['2 - Qual o grau de dificuldade desta prova na parte do Componente Específico?']
chart = pd.value_counts(s).to_frame(name='data')
ordinal_data = df.groupby('2 - Qual o grau de dificuldade desta prova na parte do Componente Específico?')[['NOTA COMPONENTE ESPECIFICO']].mean()
chart['ordinal'] = ordinal_data
chart.index.name = 'labels'
chart = chart.reset_index().sort_values('ordinal')
chart = chart[:6]
charts = [go.Bar(x=chart['labels'].values, y=chart['data'].values, name='Frequency')]
charts.append(go.Scatter(
	x=chart['labels'].values, y=chart['ordinal'].values, yaxis='y2',
	name='NOTA COMPONENTE ESPECIFICO (mean)', line={'shape': 'spline', 'smoothing': 0.3}, mode='lines'
))
figure = go.Figure(data=charts, layout=go.Layout({
    'barmode': 'group',
    'legend': {'orientation': 'h'},
    'title': {'text': '2 - Qual o grau de dificuldade desta prova na parte do Componente Específico? Value Counts'},
    'xaxis': {'title': {'text': '2 - Qual o grau de dificuldade desta prova na parte do Componente Específico?'}},
    'yaxis': {'title': {'text': 'Frequency'}},
    'yaxis2': {'overlaying': 'y', 'side': 'right', 'title': {'text': 'NOTA COMPONENTE ESPECIFICO (mean)'}}
}))

figure.show()
```
